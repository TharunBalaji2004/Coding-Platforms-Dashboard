```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector
import time

def leetcode(link):
    headers = {"User-Agent":"Mozilla/5.0"}
    r = requests.get(link, headers=headers)
    soup = bs(r.content,"html.parser")
    tags = soup.find_all("span", {"class":"mr-[5px] text-base font-medium leading-[20px] text-label-1 dark:text-dark-label-1"})
    rating = soup.find("div", {"class":"text-label-1 dark:text-dark-label-1 flex items-center text-2xl"})
    print(tags)
    if (rating == None): 
        rating = 0
    else: 
        rating = list(rating.get_text())
        rating.remove(',')
        rating = "".join(rating)

    if (tags == []):
        return '0','0','0','0','0'
    
    easy = str(tags[0].get_text())
    medium = str(tags[1].get_text())
    hard = str(tags[2].get_text())
    total = str(int(easy) + int(medium) + int(hard))

    return total, easy, medium, hard, rating

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSEA_LC WHERE REGISTERNUMBER >= '210421104054';")
results = cursor.fetchall()
count = 0
for row in results:
    if (count == 10):
        count = 0
        time.sleep(3)
    regno = row[0]
    link = row[1]
    count += 1
    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0','0','0']
    else:
        link = link.strip()
        L = leetcode(link)
    L = list(L)
    L.append(regno)

    query = "UPDATE CSEA_LC SET SOLVED={0}, EASY={1}, MEDIUM={2}, HARD={3}, RATING={4} WHERE REGISTERNUMBER={5};".format(L[0],L[1],L[2],L[3],L[4],L[5])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    print("Total problems solved: ",L[0])
    print("Easy problems solved: ",L[1])
    print("Medium problems solved: ",L[2])
    print("Hard problems solved: ",L[3])
    print("Contest Rating:",L[4])

conn.commit()
conn.close()
```

```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector
import time

def codechef(link):
    link = link.strip()
    r = requests.get(link)
    soup = bs(r.content,"html.parser")
    div_number = soup.find("div", {"class":"rating-header text-center"})

    div_number = div_number.find_all("div")
    L = []
    for tag in div_number:
        L.append(tag.get_text())
    rating = str(L[0]).strip("?i\n    Provisional Rating, click to know more")
    div = str(L[1][-2])
    star = len(str(L[2]))
    print(rating, div, star)

    return rating,div,star

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSE_GENERAL_CC;")
results = cursor.fetchall()

for row in results:
    regno = row[0]
    link = row[1]

    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0']
    else:
        link = link.strip()
        L = codechef(link)
    L = list(L)
    L.append(regno)

    query = "UPDATE CSE_GENERAL_CC SET RATING={0}, DIV_NUM={1}, STAR={2} WHERE REGISTERNUMBER={3};".format(L[0],L[1],L[2],L[3])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    conn.commit()

conn.commit()
conn.close()
```

```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector
import time

def hackerrank(link):
    link = link.strip()
    r = requests.get(link,headers={"User-Agent":"Mozilla/5.0"})
    soup = bs(r.content,"html.parser")
    badges = soup.find_all("div",attrs={"class":"hacker-badge"})
    badge_count = len(badges)
    
    python,c,cpp,java,mysql,ps = 0,0,0,0,0,0

    print("Total Badges earned: "+str(badge_count))
    for tag in badges:       
        stars = tag.find_all("svg",attrs={"class":"badge-star"})
        name = str(tag.get_text())
        count = str(len(stars))
        if ("Python" in name): python = count
        elif ("CPP" in name): cpp = count
        elif ("C" in name): c = count
        elif ("Java" in name): java = count
        elif ("Sql" in name): mysql = count
        elif ("Problem" in name): ps = count
    
    return badge_count,python,c,cpp,java,mysql,ps

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSEA_HR;")
results = cursor.fetchall()

for row in results:
    regno = row[0]
    link = row[1]

    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0','0','0','0','0']
    else:
        link = link.strip()
        L = hackerrank(link)
    L = list(L)
    L.append(regno)

    query = "UPDATE CSEA_HR SET BADGES_COUNT={0}, PYTHON={1}, C={2}, CPP={3}, JAVA={4}, MYSQL={5}, PROBLEMSOLVING={6} WHERE REGISTERNUMBER={7};".format(L[0],L[1],L[2],L[3],L[4],L[5],L[6],L[7])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    conn.commit()

conn.commit()
conn.close()
```

```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector

def geeksforgeeks(link):
    link = link.strip()
    r = requests.get(link,headers={"User-Agent":"Mozilla/5.0"})
    soup = bs(r.content,"html.parser")
    count = soup.find_all("li", {"class": "tab"})
    total = 0
    L = []
    for tags in count:
        s = tags.get_text()
        num = s[s.index('(')+1 : s.index(')')]
        L.append(str(num))
        total += int(num)
    print(L)
    if (L == []):
        return 0,0,0,0,0,0
    school,basic,easy,medium,hard = L[0],L[1],L[2],L[3],L[4]
    return str(total),school,basic,easy,medium,hard

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSEA_GFG;")
results = cursor.fetchall()

for row in results:
    regno = row[0]
    link = row[1]

    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0','0','0','0','0']
    else:
        link = link.strip()
        L = geeksforgeeks(link)
        
    L = list(L)
    L.append(regno)

    query = "UPDATE CSEA_GFG SET SOLVED={0}, SCHOOL={1}, BASIC={2}, EASY={3}, MEDIUM={4}, HARD={5} WHERE REGISTERNUMBER={6};".format(L[0],L[1],L[2],L[3],L[4],L[5],L[6])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    conn.commit()

conn.commit()
conn.close()
```
