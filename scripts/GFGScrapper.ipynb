```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector

def geeksforgeeks(link):
    link = link.strip()
    r = requests.get(link,headers={"User-Agent":"Mozilla/5.0"})
    soup = bs(r.content,"html.parser")
    count = soup.find_all("li", {"class": "tab"})
    total = 0
    L = []
    for tags in count:
        s = tags.get_text()
        num = s[s.index('(')+1 : s.index(')')]
        L.append(str(num))
        total += int(num)
    print(L)
    if (L == []):
        return 0,0,0,0,0,0
    school,basic,easy,medium,hard = L[0],L[1],L[2],L[3],L[4]
    return str(total),school,basic,easy,medium,hard

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSE_GENERAL_GFG;")
results = cursor.fetchall()

for row in results:
    regno = row[0]
    link = row[1]

    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0','0','0','0','0']
    else:
        link = link.strip()
        L = geeksforgeeks(link)
        
    L = list(L)
    L.append(regno)

    query = "UPDATE CSE_GENERAL_GFG SET SOLVED={0}, SCHOOL={1}, BASIC={2}, EASY={3}, MEDIUM={4}, HARD={5} WHERE REGISTERNUMBER={6};".format(L[0],L[1],L[2],L[3],L[4],L[5],L[6])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    conn.commit()

conn.commit()
conn.close()
```
