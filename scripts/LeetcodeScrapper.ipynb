```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector
import time

def leetcode(link):
    headers = {"User-Agent":"Mozilla/5.0"}
    r = requests.get(link, headers=headers)
    soup = bs(r.content,"html.parser")
    tags = soup.find_all("span", {"class":"mr-[5px] text-base font-medium leading-[20px] text-label-1 dark:text-dark-label-1"})
    rating = soup.find("div", {"class":"text-label-1 dark:text-dark-label-1 flex items-center text-2xl"})
    print(tags)
    if (rating == None): 
        rating = 0
    else: 
        rating = list(rating.get_text())
        rating.remove(',')
        rating = "".join(rating)

    if (tags == []):
        return '0','0','0','0','0'
    
    easy = str(tags[0].get_text())
    medium = str(tags[1].get_text())
    hard = str(tags[2].get_text())
    total = str(int(easy) + int(medium) + int(hard))

    return total, easy, medium, hard, rating

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSE_GENERAL_LC;")
results = cursor.fetchall()
count = 0
for row in results:
    if (count == 10):
        count = 0
        time.sleep(3)
    regno = row[0]
    link = row[1]
    count += 1
    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0','0','0']
    else:
        link = link.strip()
        L = leetcode(link)
    L = list(L)
    L.append(regno)

    query = "UPDATE CSEA_LC SET SOLVED={0}, EASY={1}, MEDIUM={2}, HARD={3}, RATING={4} WHERE REGISTERNUMBER={5};".format(L[0],L[1],L[2],L[3],L[4],L[5])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    print("Total problems solved: ",L[0])
    print("Easy problems solved: ",L[1])
    print("Medium problems solved: ",L[2])
    print("Hard problems solved: ",L[3])
    print("Contest Rating:",L[4])

conn.commit()
conn.close()
```
