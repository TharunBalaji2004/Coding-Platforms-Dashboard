```python
import requests
from bs4 import BeautifulSoup as bs
import mysql.connector
import time

def hackerrank(link):
    link = link.strip()
    r = requests.get(link,headers={"User-Agent":"Mozilla/5.0"})
    soup = bs(r.content,"html.parser")
    badges = soup.find_all("div",attrs={"class":"hacker-badge"})
    badge_count = len(badges)
    
    python,c,cpp,java,mysql,ps = 0,0,0,0,0,0

    print("Total Badges earned: "+str(badge_count))
    for tag in badges:       
        stars = tag.find_all("svg",attrs={"class":"badge-star"})
        name = str(tag.get_text())
        count = str(len(stars))
        if ("Python" in name): python = count
        elif ("CPP" in name): cpp = count
        elif ("C" in name): c = count
        elif ("Java" in name): java = count
        elif ("Sql" in name): mysql = count
        elif ("Problem" in name): ps = count
    
    return badge_count,python,c,cpp,java,mysql,ps

conn = mysql.connector.connect(host="localhost",user="root",passwd="562004",database="codingdashboard")
cursor = conn.cursor()
cursor.execute("SELECT REGISTERNUMBER,PROFILELINK FROM CSE_GENERAL_HR;")
results = cursor.fetchall()

for row in results:
    regno = row[0]
    link = row[1]

    print("RegNo: ",regno)
    if (len(link) == 0):
        L = ['0','0','0','0','0','0','0']
    else:
        link = link.strip()
        L = hackerrank(link)
    L = list(L)
    L.append(regno)

    query = "UPDATE CSE_GENERAL_HR SET BADGES_COUNT={0}, PYTHON={1}, C={2}, CPP={3}, JAVA={4}, MYSQL={5}, PROBLEMSOLVING={6} WHERE REGISTERNUMBER={7};".format(L[0],L[1],L[2],L[3],L[4],L[5],L[6],L[7])
    print(query)
    cursor.execute(query)
    cursor.execute("COMMIT;")
    conn.commit()

conn.commit()
conn.close()
```
